```python
#filling missing values with Mode 
data1['Area_Service'] = data1['Area_Service'].fillna(data1['Area_Service'].mode()[0])
data1['hospital_county'] = data1['hospital_county'].fillna(data1['hospital_county'].mode()[0])
#filling Hospital Id missing values with Median 
data1['hospital_id'] = data1['hospital_id'].fillna(data1['hospital_id'].median())
data1["mortality_risk"].fillna(0,inplace=True)
```
